```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv('/content/sample_data/india_electricity_consumption_data_rupees.csv')

# Step 2: Ensure the Refrigerator column is numeric (if not already)
df['Refrigerator'] = pd.to_numeric(df['Refrigerator'], errors='coerce')
df = df.drop(columns=['Refrigerator'])

# Step 3: Drop non-numeric columns before calculating correlation
numeric_columns = df.select_dtypes(include=['number'])  # Select only numeric columns

# Step 4: Calculate the correlation matrix
correlation = numeric_columns.corr()


# Step 5: Plot the heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
import pandas as pd

# Step 1: Load the dataset
df = pd.read_csv('/content/sample_data/india_electricity_consumption_data_rupees.csv')
df = df.drop(columns=['Refrigerator'])

# Step 2: Mapping month names to numbers
month_mapping = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4,
    'May': 5, 'June': 6, 'July': 7, 'August': 8,
    'September': 9, 'October': 10, 'November': 11, 'December': 12
}

# Convert the 'Month' column to numeric
df['Month'] = df['Month'].map(month_mapping)

# Step 3: Drop non-numeric columns like 'City' or 'Company'
df = df.select_dtypes(include=['number'])  # Keep only numeric columns

# Step 4: Calculate the correlation matrix for the 'ElectricityBill' column
correlation_with_electricity_bill = df.corr()['ElectricityBill']
print(correlation_with_electricity_bill)
```
